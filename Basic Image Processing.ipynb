```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread("lena.png",0)
img1=cv2.imread("lena.png",1)
img2=cv2.imread("lena.png",-1)
img=cv2.resize(img,(300,300))
img1=cv2.resize(img1,(300,300))
img2=cv2.resize(img2,(300,300))
print(cv2_imshow(img1))
img=cv2.transpose(img)
print("Transposed image",cv2_imshow(img))
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
print("Color to Grey Scale",cv2_imshow(img1))
img1=255-img1
print("Inverting color image ",cv2_imshow(img1))
img2=255+img2
print("Inverting greyscale image ",cv2_imshow(img2))
cv2.imwrite("LENA.png",img2)
```

```python
import numpy as np
new=cv2.imread("lena.png",1)
new=cv2.resize(new,(300,300))
new1=cv2.imread("lena.png",0)
new1=cv2.resize(new1,(300,300))
new2=cv2.imread("lena.png",-1)
new2=cv2.resize(new2,(300,300))
new1=cv2.cvtColor(new1,cv2.COLOR_GRAY2BGR)
temp=np.concatenate((new,new1,new2),axis=1)
print(cv2_imshow(temp))
```

```python
new=cv2.imread("lena.png",1)
new=cv2.resize(new,(300,300))
laplacian=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
temp=cv2.filter2D(src=new,ddepth=-1,kernel=laplacian)
print("Laplacian ",cv2_imshow(temp))
pre_hori=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
temp=cv2.filter2D(src=new,ddepth=-1,kernel=pre_hori)
print("Prewit Horizontal ",cv2_imshow(temp))
pre_vert=np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
temp=cv2.filter2D(src=new,ddepth=-1,kernel=pre_vert)
print("Prewit vertical ",cv2_imshow(temp))

sobel_hori=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
temp=cv2.filter2D(src=new,ddepth=-1,kernel=sobel_hori)
print("Sobel Horizontal ",cv2_imshow(temp))

sobel_vert=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
temp=cv2.filter2D(src=new,ddepth=-1,kernel=sobel_vert)
print("Sobel Vertical ",cv2_imshow(temp))

avg=np.ones((3,1))
avg/=9
temp=cv2.filter2D(src=new,ddepth=-1,kernel=avg)
print("Average ",cv2_imshow(temp))
```
